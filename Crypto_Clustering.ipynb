```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
crypto_data_scaled = StandardScaler().fit_transform(df_market_data)
crypto_data_scaled
```

```python
# Check the column names
df_market_data.columns
```

```python
# Create a DataFrame with the scaled data
crypto_data_df = pd.DataFrame(crypto_data_scaled, columns=['price_change_percentage_24h', 'price_change_percentage_7d',
       'price_change_percentage_14d', 'price_change_percentage_30d',
       'price_change_percentage_60d', 'price_change_percentage_200d',
       'price_change_percentage_1y'])
crypto_data_df

# Copy the crypto names from the original DataFrame
coin_names = list(df_market_data.index)
coin_names

# Set the coin_id column as index
crypto_data_df['coin_id']= coin_names
crypto_data_df = crypto_data_df.set_index('coin_id')

# Display the scaled DataFrame
crypto_data_df
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,12))
k
```

```python
# Create an empty list to store the inertia values
inertia_list = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(crypto_data_df)
    inertia_list.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_dict = {"k": k, "inertia": inertia_list}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_dict)
elbow_df
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
elbow_1_plot = elbow_df.hvplot.line(
    x="k",
    y="inertia",
    title= "Elbow Curve",
    xticks=k
)
elbow_1_plot
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer: Based on the elbow curve, we can see that the best value for k is 4.

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
model_2 = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the scaled DataFrame
model_2.fit(crypto_data_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
crypto_clusters = model_2.predict(crypto_data_df)

# Print the resulting array of cluster values.
crypto_clusters
```

```python
# Create a copy of the scaled DataFrame
prediction_df = crypto_data_df.copy()   
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
prediction_df['Cluster Predictions'] = crypto_clusters

# Display the copy of the scaled DataFrame
prediction_df
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
original_scaled_hover_plot =prediction_df.hvplot.scatter(
    x='price_change_percentage_24h',
    y='price_change_percentage_7d',
    by='Cluster Predictions',
    hover_cols= 'coin_id'
)
original_scaled_hover_plot
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
pca_clusters = pca.fit_transform(crypto_data_df)

# View the scaled PCA data
pca_clusters[:5]
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

```python
# Calculate the total explained variance by summing the explained variance ratios
sum(pca.explained_variance_ratio_)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer: The total explained variance of the three principal components is 0.8950316570309841. (0.3719856 + 0.34700813 + 0.17603793)

```python
# Create a new DataFrame with the PCA data.
pca_df = pd.DataFrame(
    pca_clusters,
    columns = ["PC1", "PC2", "PC3"]
)
pca_df

# Copy the crypto names from the original scaled DataFrame
pca_df['coin_id']= list(crypto_data_df.index)
pca_df

# Set the coin_id column as index
pca_df= pca_df.set_index('coin_id')

# Display the scaled PCA DataFrame
pca_df
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k_pca = list(range(1,12))
k_pca
```

```python
# Create an empty list to store the inertia values
inertia= []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=0)
    k_model.fit(pca_df)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_2 = {"k_pca": k_pca, "inertia": inertia}
# Create a DataFrame with the data to plot the Elbow curve
elbow_2_df = pd.DataFrame(elbow_2)
elbow_2_df
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
elbow_plot_pca = elbow_2_df.hvplot.line(
    x="k_pca",
    y="inertia",
    title= "Elbow Curve",
    xticks=k
)
elbow_plot_pca
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer: 4


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer: While the K value does not differ from the original data elbow curve, the inertia is lower. The PCA data has an inertia of 49.665 at a K value of 4, whereas the original data has an inertia of 79.022 at the same K value of 4. This would mean that the data points in the PCA data cluster are closer to the centroids, which makes them more compact and well defined. 

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=0)
```

```python
# Fit the K-Means model using the PCA data
model.fit(pca_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
pca_predict = model.predict(pca_df)

# Print the resulting array of cluster values.
pca_predict
```

```python
# Create a copy of the scaled PCA DataFrame
pca_predict_df= pca_df.copy()

# Add a new column to the copy of the PCA DataFrame with the predicted clusters
pca_predict_df["cluster_predictions"] = pca_predict

# Display the copy of the scaled PCA DataFrame
pca_predict_df
```

```python
# Create a scatter plot using hvPlot by setting
# `x="PC1"` and `y="PC2"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
pca_hover_plot =pca_predict_df.hvplot.scatter(
    x="PC1",
    y="PC2",
    by="cluster_predictions",
    hover_cols="coin_id"
)
pca_hover_plot
```

#### Answer the following question: 

  * **Question:** After visually analyzing the cluster analysis results, what is the impact of using fewer features to cluster the data using K-Means?

  * **Answer:** 
  Eblow Curve Comparisons: 
  Both elbow curves show the same ideal K value of 4, but the elbow curve for the PCA data shows a lower inertia value of 49 compared to 79. 

  The scaled original data uses scaling to standardize the data (ie mean=0, standard deviation= 1) and makes the influence of all features equal. The higher inertia indicates that the data has more dispersed clusters. 

  The PCA data reduces the features by focusing only on the most important pieces that explain the variance. The lower inertia would indicate that the clusters are closer and more compact. This is due to PCA filtering out less significant variance and reducing the noise.

  Scatter Cluster Comparisons:

  The first cluster graph (scaled original data) shows that the clusters are overlapping and it is difficult to see distinct clusters. This would indicate that more features being used may not be the best option since there are no clear underlying patterns or structure to the clusters.

  The second cluster graph (PCA data) shows that the clusters do not overlap as much. There are much more defined clusters- all 4 are clearly visible. This would indicate that using less features reduced the noice and gives insight to patterns in the data.
